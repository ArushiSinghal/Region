```python
import pandas as pd


def highlight_max(s):
    '''
    highlight the maximum in a Series green.
    '''
    new_data = list(s)
    max_val = 0
    for i in range(len(new_data)):
        var = new_data[i][:-1]
        var = float(var)
        if (var > max_val):
            max_val = var
    max_val = str(max_val) + "%"
    is_max = s == max_val
    return ['background-color: green' if v else '' for v in is_max]


def highlight_min(s):
    '''
    highlight the maximum in a Series green.
    '''
    new_data = list(s)
    min_val = 1000
    for i in range(len(new_data)):
        var = new_data[i][:-1]
        var = float(var)
        if (var < min_val):
            min_val = var
    min_val = str(min_val) + "%"
    
    is_min = s == min_val
    return ['background-color: yellow' if v else '' for v in is_min]

regione1= "Puglia"
download=pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-regioni/dpc-covid19-ita-regioni.csv")
print (download.shape)
print (download.columns)


```

```python
regioni= download
```

```python
temp = regioni

temp = temp.iloc[-42:]
data = temp


data["ratio"]= (data["deceduti"]/data["totale_casi"])*100
#data = data.sort_values(by='ratio', ascending=False)
data['ratio'] = data['ratio'].round(2).astype(str)+ '%'
data = data[['data','ratio','deceduti','totale_casi', 'denominazione_regione']]
print("\nHighlight the maximum value in each column:")

data.style.apply(highlight_max, subset=['ratio']).apply(highlight_min, subset=['ratio'])
```
